```python
from sklearn.datasets import load_linnerud
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = load_linnerud()

# Create DataFrames for the features and targets
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.DataFrame(dataset.target, columns=dataset.target_names)
```

```python
# Print the first few rows to verify the DataFrames
print("Features (Exercise Data):")
print(X)
```

```python
print("\nTargets (Physiological Data):")
print(y)
```

```python
# Plotting the feature set
X.plot(kind='box', subplots=True, layout=(1, 3), figsize=(12, 6), title='Exercise Data')
plt.tight_layout()
plt.show()
```

```python
# Plotting the target set
y.plot(kind='box', subplots=True, layout=(1, 3), figsize=(12, 6), title='Physiological Data')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot matrix to visualize relationships
pd.plotting.scatter_matrix(X, figsize=(12, 12), diagonal='kde')
plt.suptitle("Scatter Matrix of Exercise Data")
plt.show()

pd.plotting.scatter_matrix(y, figsize=(12, 12), diagonal='kde')
plt.suptitle("Scatter Matrix of Physiological Data")
plt.show()
```
